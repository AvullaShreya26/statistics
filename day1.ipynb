## Why missing values are replaced by central tendencies??


When a dataset has missing values, it can cause problems in analysis or machine learning because most algorithms cannot handle NaN. To solve this, we replace missing values with central tendencies like mean, median, or mode because:

Represents the typical value: Central tendencies give a value that is “average” or most common in the data, so it is a reasonable guess for the missing entry.

Prevents data loss: Instead of removing rows with missing values (which may reduce important data), we fill them to keep all rows.

Reduces bias: Using mean, median, or mode ensures the dataset remains balanced and does not distort overall patterns.

Simple and effective: It is an easy method that works well for many real-life problems.

Example:
If the Age column has missing values, replacing them with the mean age ensures those rows can still be used in analysis without affecting the overall dataset much.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r'C:\Users\Admin\Downloads\loan_approved.csv')

# Check for missing values
print(df.isnull().sum())

# Replace missing numerical values with mean (e.g., 'LoanAmount')
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)

# Replace missing numerical values with median (e.g., 'ApplicantIncome')
df['ApplicantIncome'].fillna(df['ApplicantIncome'].median(), inplace=True)

# Replace missing categorical values with mode (e.g., 'Gender')
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

# Verify missing values are handled
print(df.isnull().sum())
```
