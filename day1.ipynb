## Why missing values are replaced by central tendencies??


When a dataset has missing values, it can cause problems in analysis or machine learning because most algorithms cannot handle NaN. To solve this, we replace missing values with central tendencies like mean, median, or mode because:

Represents the typical value: Central tendencies give a value that is “average” or most common in the data, so it is a reasonable guess for the missing entry.

Prevents data loss: Instead of removing rows with missing values (which may reduce important data), we fill them to keep all rows.

Reduces bias: Using mean, median, or mode ensures the dataset remains balanced and does not distort overall patterns.

Simple and effective: It is an easy method that works well for many real-life problems.

Example:
If the Age column has missing values, replacing them with the mean age ensures those rows can still be used in analysis without affecting the overall dataset much.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r'C:\Users\Admin\Downloads\loan_approved.csv')

# Check for missing values
print(df.isnull().sum())

# Replace missing numerical values with mean (e.g., 'LoanAmount')
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)

# Replace missing numerical values with median (e.g., 'ApplicantIncome')
df['ApplicantIncome'].fillna(df['ApplicantIncome'].median(), inplace=True)

# Replace missing categorical values with mode (e.g., 'Gender')
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

# Verify missing values are handled
print(df.isnull().sum())
```

### mean, standard devition,Z-score

```python
import pandas as pd
from scipy import stats

# Load your dataset
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\loan_approved.csv')

# Iterate through numeric columns
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    col_mean = df[col].mean()
    col_std = df[col].std()
    z_scores = (df[col] - col_mean) / col_std  # manual z-score calculation
    
    print(f"Column: {col}")
    print(f"Mean: {col_mean}")
    print(f"Standard Deviation: {col_std}")
    print("Z-scores (first 5 rows):")
    print(z_scores.head())
    print("-"*50)
```

## Skewness and kurtosis were calculated for all numeric columns to understand the shape and symmetry of the data distribution.  
Based on these values, we identified which features are approximately normal and which are skewed or not normally distributed.


```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Load dataset
df = pd.read_csv("C:\\Users\\Admin\\Downloads\\loan_approved.csv")

# Select only numeric columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

print("Skewness and Kurtosis for each numeric column:\n")

for col in num_cols:
    col_skew = skew(df[col].dropna())
    col_kurt = kurtosis(df[col].dropna())
    
    print(f"Column: {col}")
    print(f"  Skewness : {col_skew}")
    print(f"  Kurtosis : {col_kurt}")

    # Normality check rules:
    # Skewness between -0.5 and +0.5 → approximately normal
    # Kurtosis near 0 → normal (excess kurtosis)
    
    if -0.5 <= col_skew <= 0.5 and -1 <= col_kurt <= 1:
        print("  🔹 Distribution: Approximately NORMAL")
    else:
        print("  Distribution: NOT NORMAL / SKEWED")

    print("-" * 50)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import PowerTransformer

# Assuming 'data' is your DataFrame and has a column 'LoanAmount'

# Initialize PowerTransformer with Yeo-Johnson method
trans = PowerTransformer(method='yeo-johnson')

# Fit and transform the 'LoanAmount' column
la_trans = trans.fit_transform(data[['LoanAmount']])
print(la_trans.shape)  # Should print (number_of_rows, 1)

# Calculate skewness
skew_value = pd.Series(la_trans.reshape(la_trans.shape[0],)).skew()
print("Skewness:", skew_value)

# Calculate kurtosis
kurt_value = pd.Series(la_trans.reshape(la_trans.shape[0],)).kurtosis()
print("Kurtosis:", kurt_value)
```

### After pt

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer

# Numerical columns
num_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']  # replace as needed

# Initialize PowerTransformer
trans = PowerTransformer(method='yeo-johnson')

# Dictionary to store transformed data and stats
transformed_data = {}
skew_kurt_stats = {}

for col in num_cols:
    # Fit and transform
    transformed = trans.fit_transform(data[[col]])
    
    # Flatten to 1D
    transformed_data[col] = transformed.flatten()
    
    # Skewness and kurtosis
    skew_val = pd.Series(transformed.flatten()).skew()
    kurt_val = pd.Series(transformed.flatten()).kurtosis()
    skew_kurt_stats[col] = {'Skewness': skew_val, 'Kurtosis': kurt_val}

# Convert to DataFrame
transformed_df = pd.DataFrame(transformed_data)

# Display skewness and kurtosis
for col, stats in skew_kurt_stats.items():
    print(f"{col} -> Skewness: {stats['Skewness']:.4f}, Kurtosis: {stats['Kurtosis']:.4f}")

# -------------------------------
# Plot histograms for comparison
# -------------------------------
plt.figure(figsize=(20, 8))

for i, col in enumerate(num_cols):
    # Original data
    plt.subplot(2, len(num_cols), i+1)
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f"{col} - Original")
    
    # Transformed data
    plt.subplot(2, len(num_cols), i+1+len(num_cols))
    sns.histplot(transformed_df[col], kde=True, color='salmon')
    plt.title(f"{col} - Transformed")

plt.tight_layout()
plt.show()
```
